## Rationale
Canopy interception is defined as rainfall which is intercepted by and evaporated from plant canopies or plant residue and is one of many processes that is involved in the soil water balance. Canopy interception can have a signicant role in reducing the amount of water that reaches the soil surface. Agricultural crops such as corn may intercept 20% to 30% of rainfall as it matures, interception values for a tallgrass praire can range from 25% upward to 60%, and a deciduous forest can intercept around 30% precipitation. Models have been developed to describe interception, however interception in turfgrass systems has not been well-defined. Interception in a turfgrass canopy occurs when precipitation is retained and redistributed to the soil (throughfall and stemflow) and evaporated from the plant leaves. Understanding the interception dynamics associated with turfgrass will allow one to make better irrigation decisions, particulary the amount of water to apply. Prelimary results have shown some turfgrass species can intercept 5 to 10 mm of rainfall.

When raindrops fall on turfgrass, water may flow through gaps in the canopy (throughfall), flow along stems to the soil surface (stemflow) or  evaporate from the surface of the canopy. The total amount of water intercepted from plant canopies can be defined as: <br/>
<br/>
 **Interception = Rainfall - (Throughfall + Stemflow)** <br/>

##### **Let's specifically look at a rainfall event period of roughly 19 hours, beginning on April-16. First, we will need to create a dataframe and specify our date range of interest.**

## **16-April-2020 Rain Event** 

```python
#import modules 
import glob 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
```

#### **Creating Dataframes**

```python
# create dataframe from csv file 
df = pd.read_csv(r"24159_Interception_2020_03_11_16_52_03.dat", skiprows=lambda x: x in [0,2,3]) #skip rows 

# specify date range 
subset_dates = pd.date_range(start ="2020-04-16 09:00:00", end = "2020-04-17 11:00:00", freq = 'min') #yyyy mm dd hour min sec 

#convert dates from index to datetime column
df["TIMESTAMP"] = pd.to_datetime(df["TIMESTAMP"], format="%Y-%m-%d %H:%M:%S")

# find matching dates 
#isin method: result will only be true if labels match
idx_subset = df["TIMESTAMP"].isin(subset_dates)

#new dataframe subset 
df_subset = df.loc[idx_subset,:,]
```

##### **From our setup design, three paired rain gauges were utilized. For simplicity, generated means from each precipitation gauge and turf patch gauge were added to the dataframe.**

```python
#precipitation gauge  
df_subset['precip_avg'] = (df_subset['Rain_mm_A_Tot'] + df_subset['Rain_mm_B_Tot'] + df_subset['Rain_mm_C_Tot']) / 3

#zoysiagrass patch gauge
df_subset['zoy_avg'] = (df_subset['Rain_mm_zoy_A_Tot'] + df_subset['Rain_mm_zoy_B_Tot'] + df_subset['Rain_mm_zoy_C_Tot']) / 3
```

##### **Let's get a feel for the data by visually ploting precipitation amounts over time. Take notice of various intensities, durations and pauses/breaks over this period.**

```python
#plot timeseries of rainfall event from df_subset
plt.figure(figsize=(12,8))

plt.plot(df_subset['TIMESTAMP'],df_subset['precip_avg']) #width=np.timedelta64(1, 'm')) #timeseries (mm-dd hour)
plt.xticks(rotation = 45, fontsize = 16)
plt.yticks(fontsize = 16)
plt.ylabel('Precipitation (mm)', fontsize = 16)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d %H:%M'))
plt.show()
```

**Figure 4**. Timeseries data depiciting precipitation amounts. 

#### **Add canopy interception metrics to dataframe**

```python
#precipitation gauge | cumulative sum; add to dataframe
df_subset['precip_cumsum'] = df_subset['precip_avg'].cumsum()

#zoysiagrass gauge | cumulative sum; add to dataframe
df_subset['zoy_cumsum'] = df_subset['zoy_avg'].cumsum()

#percentage intercepted; add to dataframe 
df_subset['pct_intercept'] = (1 - df_subset['zoy_cumsum'] / df_subset['precip_cumsum']) * 100 

#rainfall intensity | add to dataframe 
df_subset['intensity'] = df_subset['precip_avg'] * 60

#check column indexing 
#df_subset.head()
```

#### **Canopy Interception Calculations**

```python
#total precipation of rainevent 
total_precip = df_subset['precip_cumsum'].iloc[-1]

#total throughfall amount from zoysiagrass 
throughfall_zoy = df_subset['zoy_cumsum'].iloc[-1]

#total interception amount from zoysiagrass
intercept_zoy = (total_precip - throughfall_zoy)

#total precipitation amount 
pct_intercept_zoy = intercept_zoy / total_precip * 100


#for loop to calculate amount of precipiation received unitl the point of throughfall
sum_precip = 0
for i in df_subset.index:
    sum_precip = sum_precip + df_subset['precip_avg'][i] #looping and adding precipitaion values
    if df_subset['zoy_avg'][i] > 0: #checking zoysia_avg until greater than zero
        break
        
        
print(round(total_precip,2),'mm total precipitation')
print(round(throughfall_zoy,2),'mm total throughfall of zoysiagrass')
print(round(sum_precip,2), 'mm, precipitation total until the point of throughfall')
print(round(intercept_zoy, 2),'mm intercepted from zoysiagrass')
print(round(pct_intercept_zoy,2), '% intercepted from zoysiagrass')
```

```python
from pandas import Timestamp
# Generate a figure to compare cumulative precipitation and cumulative throughfall of zoysiagrass turf
plt.figure(figsize=(12,8))

plt.plot(df_subset['TIMESTAMP'],df_subset['precip_cumsum'], label = 'Precipitation')
plt.plot(df_subset['TIMESTAMP'],df_subset['zoy_cumsum'], label = 'Throughfall Zoysiagrass')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d %H:%M'))
plt.xticks(rotation = 45,fontsize=14)
plt.yticks(fontsize = 14)
plt.ylabel('Precipitation (mm)', fontsize = 14)
plt.title('Cumulative Precipitation and Throughfall', fontsize = 16)
plt.text(Timestamp("2020-04-16 09:00"), 10, "Total precipitation: 11.5mm", fontsize = 12)
plt.text(Timestamp("2020-04-16 09:00"), 9.5, "Total throughfall: 5.6mm", fontsize = 12)
plt.text(Timestamp("2020-04-16 09:00"), 9, "Interception: 5.9mm (51.3%)", fontsize = 12)
plt.text(Timestamp("2020-04-16 14:00"), 2, "Throughfall detected(3.17mm)", fontsize = 12)
plt.arrow(Timestamp("2020-04-16 14:00"),1.5, -0.04,-1, shape='full', lw=1, head_width=0.01, overhang = 10 )
plt.legend(fontsize = 12)
plt.show()
#plt.savefig("figure_5.jpg")
```

**Figure 5**. Timeseries data depiciting cumulative precipitation and throughfall.

##### **We typically think of rain events over a daily/24-hour period. However, from Figure 4, we can see that the total storm event may have multiple smaller rain events that pass through. Defining and pinpointing these events can be helpful in understanding how interception changes over time. The code below selects various rain events based on 30 minute periods without rainfall.**

```python
# storing data in rain_output
rain_output = []

without_rain_counter = 0         # time without rainfall
rain_minutes = 0                 # total minutes rainfall was measured from rain gauge
duration_minutes = 0             # total minutes of a rainfall event


#for loop that iterates over precipiation data 
for rain_row in df_subset['precip_avg']:
    duration_minutes += 1

    if float(rain_row) > 0:
        rain_minutes += 1                #if greater than zero (rain detected) then add one to rain minutes
        without_rain_counter = 0         # if greater than zero (rain detected) then without rain counter remains zero 
        
    # if the condition above is not met    
    else:
        without_rain_counter += 1        # add 1 to without rain counter (no rain detected)
 
    if without_rain_counter > 30:        # if without rain for 30 minutes (detects rainfall event) *note this value can be changed 
        if rain_minutes > 0:
            duration_minutes = duration_minutes - without_rain_counter
            rain_output.append({"Rain Minutes": rain_minutes, "Duration Minutes": duration_minutes})
                                 
        #reset counters         
        without_rain_counter = 0
        rain_minutes = 0
        duration_minutes = 0
        
    elif duration_minutes <= without_rain_counter:  # less than or equal to without rain counter
         duration_minutes = 0
                   
 #create dataframe to observe data           
df_2 = pd.DataFrame(rain_output)
df_2
```

**Dataframe Table 1**. From the this storm event we are able to detect 7 separate rainfall events based on if there is period of no rain for 30 minutes.

```python
plt.figure(figsize=(12,6))
plt.subplot()
plt.plot(df_subset['precip_cumsum'],df_subset['pct_intercept'],'k',label='Interception')
plt.ylabel('Loss of Interception, %', fontsize = 16)
plt.xlabel('Cumulative Precipitation, mm', fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend(fontsize =14)
plt.savefig('figure_6.jpg')
plt.show()
```

**Figure 6**. Loss of interception as a function of cumulative precipitation. As precipiation increases, we see a decrease in interception. Take note of the drastic decrease once precipitation values reach between 6 and 8 mm. From this entire rainfall event, 51% of the precipitation was intercepted by the zoysiagrass turf.
